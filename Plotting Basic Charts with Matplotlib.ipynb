```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Soru 1:
In this assignment, you will continue work with the Coronavirus Source Data. You will plot different chart types. Don't forget to set titles and axis labels.

Plot a bar chart for total cases of the 20 countries that have the highest numbers.

```python
covid=pd.read_csv('owid-covid-data.csv',index_col=0)

total_case=covid.groupby(by='location')['new_cases'].sum()
tc=(total_case.sort_values(ascending=False)).head(20)
locations=pd.Series(['World','United States','Brazil','India','Russia','South Africa','Peru','Mexico',
           'Colombia','Chile','Spain','Iran','United Kingdom','Saudi Arabia','Pakistan',
           'Bangladesh','Argentina','Italy','Turkey','Germany']
)


plt.figure(figsize=(12,7))
plt.xlabel('total cases')
plt.ylabel('countries')
plt.title('Total cases of the highest 20 countries')

plt.barh(locations,tc,color="#FF914D")

plt.show()
```

# Soru 2:
Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.

```python
plt.figure(figsize=(8,15))
covid=pd.read_csv('owid-covid-data.csv',index_col=0)
covid_andorra=covid[(covid['location']=='Andorra')]
covid_andorra

ax1=plt.subplot(3,1,1)
ax1.set_xlabel('new_deaths')
ax1.set_ylabel('Counts')
ax1.legend('Andorra',loc=1)
plt.hist(covid_andorra['new_deaths'], bins=20, color='g')

ax2=plt.subplot(3,1,2)
ax2.set_xlabel('new_deaths')
ax2.set_ylabel('Counts')
ax2.legend("Andorra",loc=1)
plt.hist(covid_andorra['new_deaths'], bins=50, color="r")


ax3=plt.subplot(3,1,3)
ax3.set_xlabel('new_deaths')
ax3.set_ylabel('Counts')
ax3.legend('Andorra',loc=1)
plt.hist(covid_andorra['new_deaths'], bins=10, color="b")

plt.show()
```

# Soru 3:
Plot a scatter plot of new cases and new death for Germany and France.

```python
covid=pd.read_csv('owid-covid-data.csv',index_col=0)

plt.figure(figsize=(7,4))
plt.title('New Cases & New Deaths')

for i in ['Germany','France']:
    plt.scatter(covid[covid['location']==i].new_cases,
               covid[covid['location']==i].new_deaths, label=i)
    
plt.xlabel("New Cases")
plt.ylabel("New Deaths")
plt.legend()
plt.show()
```

# Soru 5:
Calculate the total case for each continent and plot a pie chart.

```python
covid=pd.read_csv('owid-covid-data.csv',index_col=0)

total_case=covid.groupby(by=['continent'])['new_cases'].sum()
continent=['Africa','Asia','Europe','North America','Oceania','South America']
    
print(total_case)
print(continent)

plt.pie(total_case,labels=continent, autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.show()
```
