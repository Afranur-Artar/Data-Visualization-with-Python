```python
import matplotlib.pyplot as plt
import numpy as np
```

# Soru 1:
Create a new figure with matplotlib functions. Plot cosine function for the numbers between -3 π  and 3 π  with 0.1 intervals.

```python
fig, ax=plt.subplots()  # figure oluşturmak için yazılır.

# grafikte gösterilecek ve hesaplaması yapılacak olan değişkenler yazılır
x=np.arange(-3*(np.pi),3*(np.pi),0.1)
y=np.cos(x)

# fig ile tanımladığım figure hesaplanan değişkenlerin gösterilmesini sağlarız.
ax.plot(x,y,color='#22B573')
plt.show()
```
